Perform augmentation on Training dataset

```python
#zoom and horizontal flip (Or any augmentation technique of your choice)
```

```python
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Random Roation & Width Shift using Image Data Generator 
training_data_generator = ImageDataGenerator(
    rotation_range=90,
    zoom_range=0.4,
    width_shift_range=0.3,    
    fill_mode='nearest')


# Image Directory
training_image_directory = "/content/Pro-M3-Hurricane-Damage-Dataset/train"

# Generate Randomly Rotated Image files
training_augmented_images = training_data_generator.flow_from_directory(
    training_image_directory,
    target_size=(180,180))
```

Visualize the augmented images in four sub-plots

```python
#Visualize the preprocessed image
```

```python
from matplotlib import pyplot
from matplotlib.image import imread

for i in range(9):
  
	# define subplot
	pyplot.subplot(3, 3, i+1)
 
	# generate batch of images
	batch = training_augmented_images.next()
 
	# convert to unsigned integers for viewing
	image = batch[0][0].astype('uint8')
 
	# plot raw pixel data
	pyplot.imshow(image)


pyplot.show()
```
