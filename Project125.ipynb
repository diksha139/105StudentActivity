```python
#Display image
from matplotlib import pyplot
from matplotlib.image import imread

training_damaged_image = "/content/Pro-M3-Hurricane-Damage-Dataset/train/damage/image (1).jpeg"

# load image pixels
image = imread(training_damaged_image)

pyplot.title("damaged: Image 1")

# plot raw pixel data
pyplot.imshow(image)

# show the figure
pyplot.show()
```

```python
#Image Processing rotation(Student Activity starts here)
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Random Rotation using Image Data Generator 
training_data_generator = ImageDataGenerator(rotation_range=90, fill_mode='nearest')

# Image Directory
training_image_directory = "/content/Pro-M3-Hurricane-Damage-Dataset/train"

# Generate Randomly Rotated Image files
training_augmented_images = training_data_generator.flow_from_directory(training_image_directory,target_size=(180,180))
```

```python
#Visualize Preprocessed Images
from matplotlib import pyplot
from matplotlib.image import imread

for i in range(4):

	# Define a subplot
	pyplot.subplot(2, 2, i+1)
 
	# Generate batch of images
	batch = training_augmented_images.next()
 
	# Convert to unsigned integers for viewing
	image = batch[0][0].astype('uint8')
 
	# Plot raw pixel data
	pyplot.imshow(image)

# Plot the main figure(with 4 sub figures)
pyplot.show()
```

```python
#zoom and horizontal flip

import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Random Data Augmentation (Rescale, Resize, Shifts, Rotate) using ImageDataGenerator 
training_data_generator = ImageDataGenerator(zoom_range=0.3, 
                                             horizontal_flip=True, 
                                             fill_mode='nearest')

# Image Directory
training_image_directory = "/content/Pro-M3-Hurricane-Damage-Dataset/train"

# Generate Augmented Image files
training_augmented_images = training_data_generator.flow_from_directory(
                                                       training_image_directory,
                                                       target_size=(180,180))
```

```python
#Visualize Preprocessed Images
from matplotlib import pyplot
from matplotlib.image import imread

for i in range(4):

	# Define a subplot
	pyplot.subplot(2, 2, i+1)
 
	# Generate batch of images
	batch = training_augmented_images.next()
 
	# Convert to unsigned integers for viewing
	image = batch[0][0].astype('uint8')
 
	# Plot raw pixel data
	pyplot.imshow(image)

# Plot the main figure(with 4 sub figures)
pyplot.show()
```
